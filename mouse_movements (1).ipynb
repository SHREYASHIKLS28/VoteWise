```python

import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
tndata=pd.read_csv('/content/Train_Mouse.csv')
```

```python
ttdata=pd.read_csv('/content/Test_Mouse.csv')
```

```python
combined_data = pd.concat([tndata, ttdata], ignore_index=True)
```

```python
normal_events = [1,2,3]
suspicious_events = [4,5,6]
```

```python
combined_data['label'] = 0  # Initialize labels as normal (0)
combined_data.loc[combined_data['event_type'].isin(suspicious_events), 'label'] = 1  # Set labels for suspicious events as 1
```

```python
suspicious_event_detected = False
```

```python

for i, row in combined_data.iterrows():
    if row['event_type'] in suspicious_events:
        suspicious_event_detected = True
        print("Suspicious event detected: {}".format(row['event_type']))
```

```python
train_labeled = combined_data.iloc[:len(tndata)]
test_labeled = combined_data.iloc[len(tndata):]
```

```python
train_labeled.to_csv("train_labeled.csv", index=False)
test_labeled.to_csv("test_labeled.csv", index=False)
```

```python
if suspicious_event_detected:
    print("Warning: Suspicious events detected.")
```
